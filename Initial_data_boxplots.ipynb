```python
by_rep = pd.Series("repair" if (i==1) or (i==2) else "no-repair" for i in data['Participant study:'])
data['repair'] = by_rep
```

# Repair vs No-Repair

```python
for col in data.columns[2:30]:
    fig, ax = plt.subplots(figsize=(10,8))
    plt.suptitle('')
    ax.set_xlim([1, 3])
    ax.set_ylim([0, 10])
    data.boxplot(column=[col],by =data.columns[30],ax=ax, patch_artist=True)
```
