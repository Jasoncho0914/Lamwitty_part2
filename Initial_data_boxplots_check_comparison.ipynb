```python
by_rep = pd.Series("No-repair(Conditon 1 and 2)" if (i==1) or (i==2) else "Repair(Condition 3 and 4)" for i in data['Participant study:'])
data['repair'] = by_rep
```

# Sanity Check

### 'Did this agent ever try to repair a mistake it made?'
Condition 1 -> No 3

Condition 2 -> Yes 1

Condtion 3 -> No 3

Condition 4 -> Yes 1

```python
sanity_dict = {1:3,2:1,3:3,4:1}
```

```python
data_sanity_check = data[[sanity_dict[data['Participant study:'][i]] == data['Did this voice agent ever make a mistake?'][i] for i in range(0,(len(data)))]]
```

# Sanity Check 2

'Did this agent ever try to repair a mistake it made?'

C1 -> no 5

C2 -> no 5

C3 -> yes 4

C4 -> yes 4

```python
sanity_dict_2 = {1:5,2:5,3:4,4:4}
```

```python
data_sanity_check_2 = data_sanity_check[[sanity_dict_2[data_sanity_check['Participant study:'][i]] 
                                         == data_sanity_check['Did this agent ever try to repair a mistake it made?'][i] for i in data_sanity_check.index]]
```

# Repair vs No-Repair

Below are the Fifteen observations that didn't pass the check (incorrectly answering "Did this voice agent ever make a mistake?")

```python
data[[sanity_dict[data['Participant study:'][i]] != 
      data['Did this voice agent ever make a mistake?'][i] for i in range(0,(len(data)))]]
```

```python
data[[sanity_dict_2[data['Participant study:'][i]] != 
      data['Did this agent ever try to repair a mistake it made?'][i] for i in range(0,(len(data)))]]
```

### Below graph shows boxplots with and without four observations above

```python
for col in data.columns[2:30]:
    fig, ax = plt.subplots(ncols = 2, figsize=(20,8), sharey= True)
    plt.suptitle('')
    ax[0].set_xlim([1, 3])
    ax[0].set_ylim([0, 10])
    ax[1].set_xlim([1, 3])
    ax[1].set_ylim([0, 10])
    data.boxplot(column=[col],by =data.columns[30],ax=ax[0], patch_artist=True)
    data_sanity_check_2.boxplot(column=[col],by =data_sanity_check_2.columns[30],ax=ax[1], patch_artist=True)
    
```
